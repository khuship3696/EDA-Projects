Que. IPL finals venues and winners along with the number of wins

```python
final_match = match.loc[match.groupby('season')['id'].idxmax()]
venue_win = final_match.groupby(['city','winner']).size().reset_index(name='num_wins')
plt.figure(figsize= (12,6))
for team in venue_win['winner'].unique():
    team_data = venue_win[venue_win['winner'] == team]
    plt.bar(team_data['city'], team_data['num_wins'], label=team)
    
plt.xlabel('Venue')
plt.ylabel('Number of Wins')
plt.title('IPL final venues and winners along with the number of wins')
plt.legend()
```

Que. The number of seasons won by each team

```python
swon = match.groupby('winner')['season'].nunique()
print('Number of seasons won by each team:\n',swon)
plt.figure(figsize= (12,6))
swon.plot(kind='bar')

plt.xlabel('Season')
plt.ylabel('Number of seasons won')
plt.title('Number of season won by each team')
plt.legend()
```

Que. The toss winner, toss decision, winner in final matches

```python
toss_win = match.groupby(['toss_winner','toss_decision','winner']).size().reset_index(name='count')
a = toss_win.pivot_table(index='toss_winner',columns=['toss_decision'],values='count').fillna(0)

plt.figure(figsize= (12,6))
a.plot(kind='bar')
plt.xlabel('Toss Winner')
plt.ylabel('Count')
plt.title('Toss winner,Decision, and Matches Won')
plt.show()
```

Que. The man of the match for each winning team

```python
mom = match.groupby('winner')['player_of_match'].first().reset_index()
plt.figure(figsize= (12,6))
plt.barh(mom['winner'],mom['player_of_match'],color='green')

plt.xlabel('Player')
plt.ylabel('Team')
plt.title('Man of the match for each winning team')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Que. What is the probability of winning a match if the toss was won

```python
twon = match[match['toss_winner'] == match['winner']]
total_match = match.shape[0]
toss_win_match = twon.shape[0]
prob_of_winning = toss_win_match/total_match

plt.figure(figsize= (12,6))
sns.barplot(x=['Probability of winning toss and match','Probability of losing toss and winning match'],
           y=[prob_of_winning, 1- prob_of_winning])

plt.title('Probability of winning a match based on toss result')
plt.ylabel=('Probability')
plt.show()
```
